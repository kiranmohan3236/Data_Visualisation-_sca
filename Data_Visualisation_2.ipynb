### Sub Plot

```python
#half split view for better understanding
#both graph ploted half half
#plt.subplot(R,C)
plt.figure(figsize=(12,8))
plt.subplot(1,2,2)
sns.barplot(x='Genre',y='Global_Sales',data = df1,estimator=np.mean)


plt.subplot(1,2,1)
sns.boxplot(x='Publisher',y='Global_Sales',data = df1)
```

```python
#plt.subplot(R,C)
plt.figure(figsize=(12,8))
plt.subplot(3,5,3)
sns.barplot(x='Genre',y='Global_Sales',data = df1,estimator=np.mean)

plt.subplot(3,5,9)
sns.boxplot(x='Genre',y='Global_Sales',data = df1)

plt.subplot(3,5,6)
sns.countplot(data=df1,x='Genre',hue='Publisher')


plt.subplot(3,5,10)
sns.countplot(data=df1,x='Genre',hue='Publisher')


plt.subplot(3,5,12)
sns.countplot(data=df1,x='Genre',hue='Publisher')
plt.show()
```
